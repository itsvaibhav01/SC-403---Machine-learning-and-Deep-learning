```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

# Data Load

```python
training_1 = pd.read_csv('DC_Motor.csv',header=None)
training_1.head()
```

```python
training_1 = training_1.values
print(training_1.shape)
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(training_1[:,0], training_1[:,1], random_state = 1,test_size=0.3)
# X_train, X_test, y_train, y_test = training_1[:139,0], training_1[139:199,0], training_1[:139,1], training_1[139:199,1]
print(X_train.shape, X_test.shape)
# print(X_train[0:20])
# print(min(X_train),max(X_train))
```

```python
plt.plot(X_train ,'-')
plt.plot(y_train, '--')
plt.show
```

```python
plt.figure(figsize=(10,4))      
plt.subplot(1, 2, 1)
plt.plot(X_train,'--')
plt.plot(y_train,'-')
plt.title('training sequence, set ')
plt.xlabel('time')
plt.ylabel('RPM')

plt.subplot(1, 2, 2)
plt.plot(X_test,'--')
plt.plot(y_test,'-')
plt.title('test sequence, set ')
plt.xlabel('Pulse')
plt.ylabel('RPM')

plt.tight_layout()
plt.show()
```

# model 

```python
model = Sequential()
model.add(Dense(32, input_dim=1, activation='linear'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
history = model.fit(X_train, y_train,validation_split=0.33, epochs=50, batch_size=1)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('loss vs epoch ')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

# predict 

```python
y_train_predict = np.ravel(model.predict(X_train))
# y_train_predict=y_train_predict*ly_train[0]+ly_train[1] 
# X_train = X_train *lX_train[0]+lX_train[1] 
plt.plot(X_train, '--')
plt.plot(y_train_predict, '-')
plt.title('train sequence, set ')
plt.xlabel('time')
plt.ylabel('RPM')
```

```python
plt.figure(figsize=(10,5))      
y_test_predict = np.ravel(model.predict(X_test))
# y_test_predict=y_test_predict*ly_train[0]+ly_train[1] 
plt.plot(y_test_predict , 'o')
plt.plot(y_test,'x')
plt.title('real sequence vs predicted sequence')
plt.xlabel('time')
plt.ylabel('RPM')
plt.show()
print("Blue dot : Real Points ")
print("Orange cros : predicted points ")
```

```python
trainScore = (mean_squared_error(y_train_predict, y_train))
print("trainScore :",trainScore , "MSE")

testScore = (mean_squared_error(y_test_predict, y_test))
print("testScore :",testScore , "MSE")
```
